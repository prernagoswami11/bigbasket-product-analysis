# data transformation

```python
# Discount column 

data['discount'] = data['market_price'] - data['sale_price']
```

```python
# Discount Rate Column

data['Discount_rate'] = (( data['discount'] / data['market_price'])*100).round(2)
```

```python
# outliers
```

```python
below_first_quartile = data [ data['sale_price'] <= data['sale_price'].quantile(0.25) ]
```

```python
above_third_quartile = data [ data['sale_price'] >= data['sale_price'].quantile(0.75) ]
```

```python
q1 = data['sale_price'].quantile(0.25)
q3 = data['sale_price'].quantile(0.75)
```

```python
IQR = q3 - q1
IQR
```

```python
lower_limit = q1 - 1.5 * IQR
upper_limit = q3 + 1.5 * IQR
```

```python
outliers = data [  ( data['sale_price'] < lower_limit ) | ( data['sale_price'] > upper_limit ) ]
```
